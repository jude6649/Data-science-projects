**Importing Libraries**

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import StackingClassifier
import numpy as np
import pandas as pd
```

**Loading data**

```python
obesity_data=pd.read_csv('/Obesity Classification.csv')
```

# ***Showing the data***

```python
obesity_data.head()
```

```python
obesity_data.replace({'Gender':{'Male':0,'Female':1}},inplace=True)
obesity_data.replace({'Label':{'Underweight':0,'Normal Weight':1,'Overweight':2,'Obese':3}},inplace=True)


```

```python
obesity_data.head()
```

```python
obesity_data.info()
```

## ***Assigning data in to features and labels and furthur dividing into training and testing  data***

```python
x=obesity_data.drop('Label',axis=1)
y=obesity_data['Label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## ***Creating the base classifers***

```python
base_classifiers = [
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('gb', GradientBoostingClassifier(n_estimators=100, random_state=42))
]
```

## ***Creating the meta classifer***

```python
meta_classifier = LogisticRegression(random_state=42)
```

```python
stacking_r = StackingClassifier(estimators=base_classifiers, final_estimator=meta_classifier)
```

## ***Fit the Stacking Classifier on the training data***

```python
stacking_r.fit(X_train, y_train)
```

## ***Make predictions on the test set and  Calculate accuracy***

```python
y_pred = stacking_r.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Stacking Classifier Accuracy: {accuracy}")
```

## ***Predictive model***

```python
input_data=(1,25,0,175,80,25.3)
input_data_as_numpy_array=np.array(input_data)
reshaped=input_data_as_numpy_array.reshape(1,-1)
prediction=stacking_r.predict(reshaped)
print(prediction)
if prediction[0]==0:
    print('Underweight')
elif prediction[0]==1:
    print('Normal Weight')
elif prediction[0]==2:
    print('Overweight')
elif prediction[0]==3:
    print('Obese')
```
